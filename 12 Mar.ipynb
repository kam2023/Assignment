```python
# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
# of 5 using Python. Interpret the results.

import scipy.stats as stats

mean = 50
std_dev = 5
n = 100 # sample size

# calculate the standard error of the mean
sem = std_dev / (n**0.5)

# calculate the confidence interval
ci = stats.norm.interval(0.95, loc=mean, scale=sem)

print(ci)
```

## To conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution, we can use the 'scipy' library in Python. Specifically, we can use the 'chisquare()' function, which calculates the chi-square test statistic and p-value for the test.

```python
# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
# matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
# brown. Use Python to perform the test with a significance level of 0.05.

import numpy as np
from scipy.stats import chisquare

observed = np.array([48, 22, 35, 17, 18, 60]) # observed counts for blue, orange, green, yellow, red, and brown, respectively
expected = np.array([60, 60, 60, 30, 30, 60]) # expected counts for blue, orange, green, yellow, red, and brown, respectively

chi2, p_value = chisquare(observed, f_exp=expected)

print("Chi-square test statistic:", chi2)
print("P-value:", p_value)

if p_value < 0.05:
    print("We reject the null hypothesis that the observed distribution matches the expected distribution.")
else:
    print("We fail to reject the null hypothesis that the observed distribution matches the expected distribution.")
```

```python
# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
# smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
#population who smoke.

from statsmodels.stats.proportion import proportion_confint

n = 500 # sample size
p = 60 / n # sample proportion

# calculate the confidence interval
ci = proportion_confint(count=60, nobs=n, alpha=0.05, method='normal')

print(ci)
```

```python
# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
# of 12 using Python. Interpret the results.

import scipy.stats as stats

sample_mean = 75
sample_std = 12
n = 100 # sample size
alpha = 0.1 # 90% confidence level

# calculate the confidence interval
ci = stats.t.interval(alpha=alpha, df=n-1, loc=sample_mean, scale=sample_std/np.sqrt(n))

print(ci)
```

```python
# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
# area corresponding to a chi-square statistic of 15.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10 # degrees of freedom
x = np.linspace(0, 30, 500) # generate x values
y = chi2.pdf(x, df) # calculate y values for the PDF

# plot the chi-square distribution
plt.plot(x, y)

# shade the area corresponding to a chi-square statistic of 15
mask = x > 15
plt.fill_between(x[mask], y[mask], color='red')

# add labels and title
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')

plt.show()
```

```python
# Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
# preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
# prefer Coke.

import numpy as np
import scipy.stats as stats

n = 1000 # sample size
p = 520/n # sample proportion
confidence_level = 0.99 # desired confidence level
z_score = stats.norm.ppf((1+confidence_level)/2) # z-score for the desired confidence level

margin_of_error = z_score * np.sqrt((p*(1-p))/n)
lower_ci = p - margin_of_error
upper_ci = p + margin_of_error

print("99% confidence interval: ({:.4f}, {:.4f})".format(lower_ci, upper_ci))

```

# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
# 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
# expected frequencies of a fair coin. Use a significance level of 0.05.

## To conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin, we need to follow these steps:

1. State the null hypothesis (H0) and alternative hypothesis (Ha)
2. Choose the significance level (α)
3. Determine the expected frequencies under the null hypothesis
4. Calculate the chi-square statistic
5. Determine the degrees of freedom
6. Calculate the p-value
7. Make a decision based on the p-value and the significance level

## Step 1: The null hypothesis is that the coin is fair, with an equal probability of heads and tails. The alternative hypothesis is that the coin is biased towards tails.

H0: The coin is fair (p = 0.5)
Ha: The coin is biased towards tails (p < 0.5)

Step 2: We choose a significance level of 0.05.

Step 3: Under the null hypothesis, we expect the number of tails to be 50 and the number of heads to be 50.

Step 4: To calculate the chi-square statistic, we use the formula:

χ2 = ∑(O - E)² / E

where O is the observed frequency and E is the expected frequency.

```python
# We can calculate the chi-square statistic using Python as follows:

import numpy as np
import scipy.stats as stats

observed = np.array([55, 45]) # tails, heads
expected = np.array([50, 50]) # equal probability of tails and heads

chi2_statistic = np.sum((observed - expected)**2 / expected)
print("Chi-square statistic:", chi2_statistic)
```

## Step 5: The degrees of freedom for a goodness of fit test with two categories is (number of categories - 1) = 1.

## Step 6: To calculate the p-value, we use the chi-square distribution with 1 degree of freedom and the cumulative distribution function (CDF) of the chi-square distribution. We want to find the probability of getting a chi-square statistic as extreme as or more extreme than the observed value (1.0) under the null hypothesis.

```python
# We can calculate the p-value using Python as follows:

p_value = 1 - stats.chi2.cdf(chi2_statistic, df=1)
print("p-value:", p_value)
```

## Step 7: Since the p-value (0.317) is greater than the significance level (0.05), we fail to reject the null hypothesis. This means that we do not have sufficient evidence to conclude that the coin is biased towards tails. The observed frequencies do not significantly differ from the expected frequencies of a fair coin at the 5% significance level.

+ Therefore, we can conclude that there is no evidence to support the hypothesis that the coin is biased towards tails based on the results of the chi-square goodness of fit test.

## Q9. To conduct a chi-square test for independence, we first need to state the null and alternative hypotheses:

+ Null hypothesis: There is no association between smoking status and lung cancer diagnosis.
+ Alternative hypothesis: There is a significant association between smoking status and lung cancer diagnosis.

## We can use the following steps to perform the chi-square test:

1. Calculate the expected frequencies for each cell assuming no association between smoking status and lung cancer diagnosis.
2. Calculate the chi-square test statistic using the observed and expected frequencies.
3. Calculate the p-value associated with the chi-square test statistic.
4. Compare the p-value to the significance level and make a conclusion about the null hypothesis.

```python
# Q 9. Edited; We can use the following steps to perform the chi-square test:

import numpy as np
from scipy.stats import chi2_contingency

# Set up the contingency table
observed = np.array([[200, 50], [100, 150]])

# Calculate the expected frequencies
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
expected = np.outer(row_totals, col_totals) / observed.sum()

# Calculate the chi-square test statistic and p-value
chi2, p_val, dof, expected = chi2_contingency(observed)

# Output the results
print("Chi-square test statistic:", chi2)
print("p-value:", p_val)

```

# Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
# deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
# Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
# significance level of 0.05.

# To conduct a hypothesis test for a population mean, we first need to state the null and alternative hypotheses:

+ Null hypothesis: The population mean is not significantly different from 70.
+ Alternative hypothesis: The population mean is significantly different from 70.

```python
# We can use a t-test to test this hypothesis. Since the population standard deviation is unknown, we will use 
# the t-distribution and calculate  the t-statistic and p-value:

import numpy as np
from scipy.stats import t

# Set the sample size and sample mean
n = 30
sample_mean = 72

# Set the null hypothesis value and sample standard deviation
null_value = 70
sample_std = 10

# Calculate the t-statistic
t_stat = (sample_mean - null_value) / (sample_std / np.sqrt(n))

# Calculate the p-value
p_val = 2 * (1 - t.cdf(abs(t_stat), n-1))

# Output the results
print("t-statistic:", t_stat)
print("p-value:", p_val)
```
