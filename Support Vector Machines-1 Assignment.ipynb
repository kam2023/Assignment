# Q1. What is the mathematical formula for a linear SVM?

+ A linear Support Vector Machine (SVM) is a classification algorithm that finds a hyperplane that best separates two classes in a feature space. The mathematical formula for a linear SVM can be represented as follows:

+ Given a set of training data with input vectors x_i ∈ R^n and binary class labels y_i ∈ {-1, 1}, i=1,...,m, the goal of linear SVM is to find a hyperplane represented as w^T x + b = 0 that maximally separates the two classes. Here, w is the weight vector and b is the bias term.

+ The optimization problem for the linear SVM can be written as follows:

+ minimize 1/2 * ||w||^2
+ subject to y_i (w^T x_i + b) >= 1 for all i

+ Here, the ||w||^2 represents the norm of the weight vector w and the term y_i (w^T x_i + b) represents the margin, which is the distance between the hyperplane and the closest data points from each class.

+ The objective of this optimization problem is to minimize the norm of the weight vector w while ensuring that all training points are correctly classified and lie at least at a distance of 1/||w|| from the hyperplane.

+ Once the optimization problem is solved, the sign of the decision function f(x) = w^T x + b is used to classify new data points. If f(x) is positive, the point is classified as belonging to the positive class (+1), and if f(x) is negative, the point is classified as belonging to the negative class (-1).

# Q2. What is the objective function of a linear SVM?

+ The objective function of a linear Support Vector Machine (SVM) is to find a hyperplane that best separates two classes in a feature space. The SVM achieves this by maximizing the margin, which is the distance between the hyperplane and the closest data points from each class.

+ The objective function of a linear SVM can be expressed as an optimization problem with the following objective:

minimize 1/2 * ||w||^2

+ subject to y_i (w^T x_i + b) >= 1 for all i

+ Here, ||w||^2 represents the squared Euclidean norm of the weight vector w, and y_i is the binary class label of the i-th training example, which can be either -1 or +1. The term (w^T x_i + b) represents the decision function, which is positive for data points that are classified as belonging to the positive class and negative for data points that are classified as belonging to the negative class.

+ The optimization problem minimizes the norm of the weight vector w while ensuring that all training points are correctly classified and lie at least at a distance of 1/||w|| from the hyperplane. This is achieved by constraining the decision function to be greater than or equal to 1 for all training examples.

+ The objective of the linear SVM is to find the optimal values of the weight vector w and bias term b that satisfy the above constraints and maximize the margin between the hyperplane and the closest data points from each class. This is achieved by solving the above optimization problem using techniques such as quadratic programming or gradient descent.

# Q3. What is the kernel trick in SVM?


+ The kernel trick is a technique used in Support Vector Machine (SVM) algorithms to transform the input data into a higher-dimensional feature space, where it may be more easily separable by a hyperplane.

+ In traditional linear SVM, the algorithm seeks to find a linear decision boundary in the original input space. However, in many cases, the data may not be linearly separable in the input space, and finding a linear decision boundary may be impossible or suboptimal.

+ The kernel trick overcomes this limitation by implicitly mapping the input data into a higher-dimensional space using a kernel function, which is a mathematical function that measures the similarity between any two data points in the input space. The kernel function calculates the dot product between the transformed input data points, which can be computed efficiently using kernel matrices, without actually computing the transformed features themselves.

+ This kernel trick allows SVM to find a non-linear decision boundary in the higher-dimensional feature space, without actually having to explicitly transform the data. This approach is computationally efficient and avoids the "curse of dimensionality" problem that occurs when explicitly transforming the data into a higher-dimensional space.

+ There are several commonly used kernel functions in SVM, including the linear kernel, polynomial kernel, Gaussian (or radial basis function) kernel, and sigmoid kernel. The choice of kernel function depends on the characteristics of the data and the problem being solved.

# Q4. What is the role of support vectors in SVM Explain with example

### Support vectors are the data points that lie closest to the decision boundary in a Support Vector Machine (SVM). These points are crucial in determining the position of the decision boundary, which in turn defines the classification boundary between two classes.

+ In SVM, the goal is to find a hyperplane that separates the two classes with the maximum margin. The margin is the distance between the hyperplane and the closest data points from each class. The data points that are closest to the hyperplane and lie on the margin are called support vectors.

```python
# Python code that demonstrates how to use SVM with a non-linearly separable dataset:
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create a non-linearly separable dataset
X = np.array([[1, 2], [2, 3], [3, 1], [4, 2], [2, 1], [3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# Define the SVM model
model = svm.SVC(kernel='rbf', gamma=1)

# Train the model
model.fit(X, y)

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
plt.show()
```

+ In this code, we first create a non-linearly separable dataset with two classes, represented by the arrays 'X' and 'y'. We then define an SVM model with an RBF (Radial Basis Function) kernel and a gamma parameter of 1. We then train the model on the dataset using the 'fit()' method.

+ Next, we plot the decision boundary of the SVM model using the 'contourf()' function. The 'predict()' method of the SVM model is used to predict the class of each point in a grid of points, which are then plotted as a contour plot. We also plot the original data points using the 'scatter()' function.

+ When you run this code, you should see a plot with the decision boundary of the SVM model separating the two classes. Note that since the dataset is not linearly separable, the SVM model has mapped the data into a higher-dimensional space using the RBF kernel function to find a hyperplane that separates the two classes with the maximum margin.

+ The role of support vectors in SVM is to define the position of the hyperplane and the margin. Changing the position of any other data point in the dataset would not affect the hyperplane or the margin. Therefore, support vectors are critical in determining the decision boundary.

+ During training, the SVM algorithm identifies the support vectors and uses them to construct the decision boundary. During testing, the algorithm classifies new data points by determining which side of the decision boundary they fall on.

+ In summary, support vectors are the data points that lie closest to the decision boundary in SVM. They play a critical role in determining the position of the decision boundary, which defines the classification boundary between two classes.

```python
# Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

#  Illustrate with examples and graphs of Hyperplane

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Define the dataset
X = np.array([[1, 2], [2, 3], [3, 1], [4, 2], [2, 1], [3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# Create the SVM model with a linear kernel
clf = svm.SVC(kernel='linear', C=1)

# Fit the model to the data
clf.fit(X, y)

# Get the separating hyperplane
w = clf.coef_[0]
b = clf.intercept_[0]
x_min = np.min(X[:, 0])
x_max = np.max(X[:, 0])
y_min = np.min(X[:, 1])
y_max = np.max(X[:, 1])
xx = np.linspace(x_min, x_max)
yy = (-w[0] * xx - b) / w[1]

# Plot the data and the hyperplane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.plot(xx, yy, 'k-')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Hyperplane Illustration')
plt.show()
```

```python
#  Illustrate with examples and graphs of Marginal plane

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Define the dataset
X = np.array([[1, 2], [2, 3], [3, 1], [4, 2], [2, 1], [3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# Create the SVM model with a linear kernel and a large C value to enforce a soft margin
clf = svm.SVC(kernel='linear', C=1000)

# Fit the model to the data
clf.fit(X, y)

# Get the support vectors and their coefficients
sv = clf.support_vectors_
alpha = clf.dual_coef_

# Get the intercept of the marginal plane
b = clf.intercept_

# Define the range of X1 and X2 values
x_min = np.min(X[:, 0])
x_max = np.max(X[:, 0])
y_min = np.min(X[:, 1])
y_max = np.max(X[:, 1])

# Define a set of points in the X1-X2 plane
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
xy = np.column_stack([xx.ravel(), yy.ravel()])

# Calculate the decision function on the set of points
Z = clf.decision_function(xy)
Z = Z.reshape(xx.shape)

# Calculate the upper and lower margins
upper_margin = Z + 1 / np.sqrt(np.sum(clf.coef_ ** 2))
lower_margin = Z - 1 / np.sqrt(np.sum(clf.coef_ ** 2))

# Plot the data, support vectors, and marginal plane
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.scatter(sv[:, 0], sv[:, 1], s=100, facecolors='none', edgecolors='k')
plt.contour(xx, yy, Z, levels=[0], linestyles=['-'])
plt.contour(xx, yy, upper_margin, levels=[0], linestyles=['--'])
plt.contour(xx, yy, lower_margin, levels=[0], linestyles=['--'])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Marginal Plane Illustration')
plt.show()
```

```python
# Illustrate with examples and graphs of Soft margin

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# generate some random data with two classes
X, y = make_blobs(n_samples=100, centers=2, random_state=42)

# create a soft margin SVM classifier
clf = SVC(kernel='linear', C=0.5)
clf.fit(X, y)

# plot the data and the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create a meshgrid to plot the decision boundary
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot the decision boundary and the margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
# Try different values of the regularisation parameter C and see how it affects the performance of the model.

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()

# Select two of the features (sepal length and petal width) for visualization
X = iris.data[:, [0, 3]]
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Try different values of the regularization parameter C
C_values = [0.001, 0.01, 0.1, 1, 10, 100]
accuracies = []
for C in C_values:
    # Train a linear SVM classifier on the training set with the current value of C
    svm_clf = LinearSVC(C=C, random_state=42)
    svm_clf.fit(X_train, y_train)

    # Predict the labels for the testing set
    y_pred = svm_clf.predict(X_test)

    # Calculate the accuracy of the classifier
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the accuracies as a function of C
plt.plot(C_values, accuracies, 'o-')
plt.xscale('log')
plt.xlabel('Regularization parameter C')
plt.ylabel('Accuracy')
plt.title('Effect of regularization parameter on SVM accuracy')
plt.show()
```

```python
# Bonus task: Implement a linear SVM classifier from scratch using Python and compare its
# performance with the scikit-learn implementation.

import numpy as np

class LinearSVM:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias to zeros
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        for _ in range(self.n_iters):
            # Select a random sample
            i = np.random.randint(n_samples)
            xi = X[i]
            yi = y[i]

            # Calculate the margin
            margin = yi * (np.dot(xi, self.weights) + self.bias)

            # If margin is less than 1, update the weights and bias
            if margin < 1:
                self.weights -= self.learning_rate * (-2 * 0.1 * self.weights + yi * xi)
                self.bias -= self.learning_rate * yi

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.sign(linear_output)

    def score(self, X, y):
        y_pred = self.predict(X)
        accuracy = np.sum(y_pred == y) / len(y)
        return accuracy
```
