# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

## Estimation statistics is a branch of statistics that deals with estimating the value of a population parameter from a sample of data.
+ There are two main types of estimates in estimation statistics: point estimates and interval estimates.

1. Point estimate: A point estimate is a single numerical value that is used to estimate the value of a population parameter. For example, the sample mean is often used as a point estimate of the population mean.

2. Interval estimate: An interval estimate is a range of values that is used to estimate the value of a population parameter. This range is typically expressed as a confidence interval, which is a range of values that is likely to contain the true value of the population parameter with a certain level of confidence. For example, a 95% confidence interval for the population mean is a range of values that is expected to contain the true population mean with a confidence level of 95%.

+ Interval estimates are more informative than point estimates because they provide a range of plausible values for the population parameter, rather than a single value. However, interval estimates require more calculations and assumptions than point estimates, so they are typically more complicated to calculate and interpret.

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

def estimate_population_mean(sample_mean, sample_std, sample_size):
    import math
    # calculate the standard error of the mean
    std_error = sample_std / math.sqrt(sample_size)
    # calculate the margin of error (for 95% confidence interval)
    margin_error = 1.96 * std_error
    # calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    # return the confidence interval
    return (lower_bound, upper_bound)
```

```python
# To use this function, you can pass in the sample mean, sample standard deviation, and sample size as arguments, like this:

sample_mean = 10.2
sample_std = 2.3
sample_size = 50

population_mean = estimate_population_mean(sample_mean, sample_std, sample_size)

print("95% confidence interval for population mean:", population_mean)
```

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

+ Hypothesis testing is a statistical method used to evaluate whether a hypothesis about a population parameter is supported by the evidence from a sample of data. It involves making a statistical inference about the population parameter based on the sample data and evaluating whether the sample evidence is strong enough to reject or fail to reject the null hypothesis.

+ The null hypothesis is a statement that assumes that there is no significant difference between the observed sample data and the hypothesized population parameter, while the alternative hypothesis is a statement that assumes that there is a significant difference between the observed sample data and the hypothesized population parameter.

+ Hypothesis testing is used in various fields of research, including medicine, psychology, economics, and engineering, to test theories and make inferences about populations based on sample data.

+ The importance of hypothesis testing lies in its ability to provide a rigorous framework for evaluating the evidence for or against a hypothesis. It allows researchers to make informed decisions based on statistical evidence, rather than relying on intuition or anecdotal evidence. By using hypothesis testing, researchers can identify patterns in data that may be due to chance or random variation, as opposed to actual differences in the population. This helps to increase the reliability and validity of research findings.

+ Hypothesis testing also provides a means of testing new ideas and theories, and can help to generate new research questions and hypotheses. It is an essential tool in the scientific method and plays a critical role in advancing knowledge in various fields of study.

# Q4. Create a hypothesis that states whether the average weight of male college students is greater than 
# the average weight of female college students.

+ Null Hypothesis: The average weight of male college students is equal to or less than the average weight of female college students.

+ Alternative Hypothesis: The average weight of male college students is greater than the average weight of female college students.

+ Symbolically, we can represent these hypotheses as follows:

H0: μmale ≤ μfemale

Ha: μmale > μfemale

+ where H0 represents the null hypothesis, Ha represents the alternative hypothesis, μmale represents the population mean weight of male college students, and μfemale represents the population mean weight of female college students.

```python
#Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.

import numpy as np
from scipy.stats import t

# Set the significance level
alpha = 0.05

# Define the two samples
sample1 = np.array([10, 12, 14, 16, 18])
sample2 = np.array([8, 11, 13, 15, 17])

# Calculate the sample means
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)

# Calculate the sample standard deviations
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the standard error of the difference between the means
std_error = np.sqrt((std1**2 / len(sample1)) + (std2**2 / len(sample2)))

# Calculate the t-statistic
t_stat = (mean1 - mean2) / std_error

# Calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# Calculate the critical value for the t-distribution
t_crit = t.ppf(1 - alpha, df)

# Compare the t-statistic to the critical value
if t_stat > t_crit:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

# Q7: Write down the steps involved in hypothesis testing.
##  the general steps involved in hypothesis testing:

1. State the null and alternative hypotheses: The first step is to clearly state the null and alternative hypotheses in terms of the population parameters being tested.

2. Set the level of significance: Determine the level of significance or the alpha level, which is the probability of rejecting the null hypothesis when it is actually true. The most commonly used level of significance is 0.05.

3. Select the test statistic: Choose the appropriate test statistic based on the type of data and the research question being investigated.

4. Determine the critical value: Calculate the critical value based on the chosen level of significance and the degrees of freedom.

5. Collect data and calculate test statistic: Collect a sample of data from the population and calculate the test statistic based on the sample data.

6. Determine the p-value: Calculate the p-value, which is the probability of obtaining the observed sample results or more extreme results assuming the null hypothesis is true.

7. Make a decision: Compare the p-value to the level of significance. If the p-value is less than the level of significance, reject the null hypothesis. If the p-value is greater than the level of significance, fail to reject the null hypothesis.

8. Interpret the results: Interpret the results of the hypothesis test in the context of the research question and draw conclusions based on the decision made in step 7.

9. Communicate the results: Present the results of the hypothesis test in a clear and concise manner, including the decision made and the reasoning behind it, along with any limitations and recommendations for future research.

+ It's important to note that the specific steps involved in hypothesis testing can vary depending on the type of test being used and the nature of the research question. However, these general steps provide a framework for conducting hypothesis tests.

# Q8. Define p-value and explain its significance in hypothesis testing.

+ In hypothesis testing, the p-value (short for "probability value") is the probability of obtaining a test statistic as extreme as or more extreme than the observed results, assuming that the null hypothesis is true. In other words, it tells you how likely it is that the observed results occurred by chance alone, given the null hypothesis.

+ The significance of the p-value lies in its use as a threshold for making decisions about whether to reject or fail to reject the null hypothesis. If the p-value is less than the level of significance (usually set at 0.05), then the observed results are considered statistically significant, and the null hypothesis is rejected in favor of the alternative hypothesis. This means that the observed results are unlikely to have occurred by chance alone and that there is strong evidence to support the alternative hypothesis.

+ On the other hand, if the p-value is greater than the level of significance, then the observed results are not considered statistically significant, and the null hypothesis is not rejected. This means that there is not enough evidence to support the alternative hypothesis, and the observed results are considered to be consistent with what would be expected if the null hypothesis were true.

+ In summary, the p-value provides a way to quantify the strength of evidence against the null hypothesis and helps researchers make decisions about whether to accept or reject their hypothesis. It is an important tool for statistical inference and helps to ensure that scientific research is conducted in a rigorous and objective manner.

```python
# Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
# parameter set to 10.

import numpy as np
import matplotlib.pyplot as plt

# Set degrees of freedom
df = 10

# Generate x values for the t-distribution
x = np.linspace(-5, 5, 500)
t = (np.sqrt(df) * x) / np.sqrt(df + x**2)

# Calculate y values for the t-distribution
y = (1 + (t**2/df))**(-(df+1)/2) / (np.sqrt(df) * np.abs(t))

# Create the plot
plt.plot(x, y)
plt.title("Student's t-Distribution with 10 Degrees of Freedom")
plt.xlabel("x")
plt.ylabel("Probability density function")
plt.show()
```

```python
#Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.

import numpy as np
from scipy.stats import t

# Define two random samples of equal size
sample1 = np.array([12, 14, 10, 11, 13])
sample2 = np.array([15, 16, 18, 12, 17])

# Calculate the mean and standard deviation of each sample
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((len(sample1)-1)*std1**2 + (len(sample2)-1)*std2**2) / (len(sample1) + len(sample2) - 2))

# Calculate the t-statistic
t_stat = (mean1 - mean2) / (pooled_std * np.sqrt(1/len(sample1) + 1/len(sample2)))

# Calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# Calculate the p-value
p_value = 2 * (1 - t.cdf(abs(t_stat), df=df))

# Set the level of significance
alpha = 0.05

# Compare the p-value to the level of significance
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
    
# Print the results
print("Sample 1 mean: {:.2f}".format(mean1))
print("Sample 2 mean: {:.2f}".format(mean2))
print("Pooled standard deviation: {:.2f}".format(pooled_std))
print("T-statistic: {:.2f}".format(t_stat))
print("P-value: {:.2f}".format(p_value))
```

# Q11: What is Student’s t distribution? When to use the t-Distribution.

+ Student's t-distribution is a probability distribution that is used in statistical inference for hypothesis testing and confidence interval estimation when the sample size is small or the population variance is unknown. It is a bell-shaped distribution similar to the standard normal distribution but with heavier tails, meaning that extreme values are more likely to occur than in a normal distribution.

## The t-distribution is used when the following conditions are met:

1. The population standard deviation is unknown and must be estimated from the sample.
2. The sample size is small (typically less than 30) and the population is not normally distributed.
3. The test statistic follows a t-distribution when the null hypothesis is true.

+ The t-distribution is commonly used in hypothesis testing to determine whether a sample mean is significantly different from a hypothesized population mean. It is also used in confidence interval estimation to determine a range of values that the population mean is likely to fall within, based on the sample mean and standard deviation.

+ Overall, the t-distribution is a valuable tool in statistics for situations where the sample size is small or the population standard deviation is unknown, providing a more accurate estimate of the true population mean or difference between means.

# Q12: What is t-statistic? State the formula for t-statistic.

+ The t-statistic is a measure of how many standard errors the sample mean is away from the hypothesized population mean. It is used in hypothesis testing to determine whether the difference between the sample mean and the hypothesized population mean is statistically significant.

## The formula for the t-statistic is:

+ t = (x̄ - μ) / (s / √n)

+ where:

* x̄ is the sample mean
* μ is the hypothesized population mean
* s is the sample standard deviation
* n is the sample size

+ The t-statistic represents the number of standard deviations that the sample mean is away from the hypothesized population mean, taking into account the variability in the sample data. The larger the absolute value of the t-statistic, the further away the sample mean is from the hypothesized population mean, indicating stronger evidence against the null hypothesis.

+ The t-statistic follows a t-distribution with n-1 degrees of freedom, where n is the sample size. The t-distribution is similar to the standard normal distribution but with heavier tails, which is why it is used when the population standard deviation is unknown and must be estimated from the sample.

# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a

H0: μ = 10 (the drug does not decrease blood pressure by 10 mmHg)
Ha: μ < 10 (the drug decreases blood pressure by more than 10 mmHg)

+ We will use a significance level of 0.05, which means that we are willing to accept a 5% chance of making a type I error (rejecting the null hypothesis when it is actually true).

+ The test statistic for a one-sample t-test is:

+ t = (x̄ - μ) / (s/√n)

+ where x̄ is the sample mean, μ is the hypothesized population mean, s is the sample standard deviation, and n is the sample size.

+ Plugging in the values, we get:

+ t = (8 - 10) / (3/√100)
+ t = -2 / 0.3
+ t = -6.67

+ The critical t-value for a one-tailed test with 99 degrees of freedom (n-1) and a 0.05 significance level is approximately -1.66 (found using a t-distribution table or calculator).

+ Since our calculated t-value of -6.67 is less than the critical t-value of -1.66, we can reject the null hypothesis and conclude that there is sufficient evidence to support the alternative hypothesis that the drug decreases blood pressure by more than 10 mmHg.

+ In other words, the sample mean decrease in blood pressure of 8 mmHg is significantly lower than the hypothesized decrease of 10 mmHg, with a p-value of less than 0.05.

Ha: μ1 ≠ μ2 (the population means for the two groups are not equal)

+ We will use a significance level of 0.01, which means that we are willing to accept a 1% chance of making a type I error (rejecting the null hypothesis when it is actually true).

+ The test statistic for a two-sample t-test is:

+ t = (x̄1 - x̄2) / (s_p * √(1/n1 + 1/n2))

+ where x̄1 and x̄2 are the sample means, s_p is the pooled standard deviation, and n1 and n2 are the sample sizes.

+ The pooled standard deviation is calculated as:

+ s_p = sqrt(((n1-1)*s1^2 + (n2-1)*s2^2) / (n1+n2-2))

+ where s1 and s2 are the sample standard deviations.

+ Plugging in the values, we get:

+ s_p = sqrt(((30-1)*10^2 + (40-1)*8^2) / (30+40-2))
+ s_p = 9.01

+ t = (80-75) / (9.01 * √(1/30 + 1/40))
+ t = 3.13

+ The critical t-value for a two-tailed test with 68 degrees of freedom (n1+n2-2) and a 0.01 significance level is approximately ±2.601 (found using a t-distribution table or calculator).

+ Since our calculated t-value of 3.13 is greater than the critical t-value of ±2.601, we can reject the null hypothesis and conclude that there is sufficient evidence to support the alternative hypothesis that the population means for the two groups are not equal.

# Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV

Sample size n = 50

Sample mean x̄ = 4

Sample standard deviation s = 1.5

Confidence level = 99%

The formula for calculating the confidence interval is:

Confidence interval = sample mean ± (t-value * standard error)

where standard error = s/√n

The t-value can be obtained from the t-distribution table using n - 1 degrees of freedom and the confidence level.

Using a t-distribution table for 49 degrees of freedom and a confidence level of 99%, the t-value is 2.678.

Substituting the values in the formula, we get:

Standard error = 1.5/√50 = 0.212

Confidence interval = 4 ± (2.678 * 0.212) = [3.425, 4.575]

Therefore, we can say with 99% confidence that the average number of ads watched by viewers during a TV program is between 3.425 and 4.575.
