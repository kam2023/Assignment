# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
 ## Matplotlib is a Python library used for creating visualizations such as graphs, charts, and plots. It is a popular library among data scientists, engineers, and researchers for data visualization purposes. Matplotlib provides a wide range of customizable plots, and it is highly extensible and easily integrated into Python applications.
 
 ### Here are five plots that can be created using the Pyplot module of Matplotlib:
 
 1. Line plot: A line plot is a type of plot that displays data points as a series of connected lines. It is used to visualize trends or changes in data over time.

2. Bar plot: A bar plot is a type of plot that displays data as rectangular bars. It is commonly used to compare the values of different categories.

3. Scatter plot: A scatter plot is a type of plot that displays data as a collection of points. It is used to visualize the relationship between two variables.

4. Histogram: A histogram is a type of plot that displays the distribution of a dataset. It is commonly used to show the frequency distribution of a variable.

5. Pie chart: A pie chart is a type of plot that displays data as slices of a circle. It is used to represent the proportions of different categories in a dataset.

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
# A scatter plot is a type of plot that displays data as a collection of points. It is commonly used to visualize the relationship between two variables.
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter plot of x and y")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()

```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

# The 'subplot()' function in Matplotlib is used to create multiple plots in a single figure. It takes three arguments: 
# the number of rows of plots, the number of columns of plots, and the index of the current plot being created. 
# The index starts at 1 and increases row-wise.
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line plot 1")

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line plot 2")

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line plot 3")

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line plot 4")

plt.suptitle("Four line plots using subplot()")
plt.show()

```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

## A bar plot is a type of plot that displays data as rectangular bars, where the length of each bar is proportional 
## to the value being represented. It is commonly used to compare the relative sizes or 
## values of different categories or groups.

import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.bar(company, profit)
plt.title("Vertical bar plot of company profits")
plt.xlabel("Company")
plt.ylabel("Profit (in billions)")

plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Horizontal bar plot of company profits")
plt.xlabel("Profit (in billions)")
plt.ylabel("Company")

plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

##  A box plot is a type of plot that is used to display the distribution of a dataset. It is also known as a
## box-and-whisker plot because it shows the box, which represents the interquartile range (IQR) of the data, and 
## the whiskers, which represent the rest of the distribution. Box plots can be useful for 
## identifying outliers and comparing the distributions of different datasets.

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the two datasets into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)

# Add title and labels for the axes
plt.title("Box plot of two datasets")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel("Values")

plt.show()
```

## In this code, we first import the necessary libraries and define the two datasets, 'box1' and 'box2'. We then combine the two datasets into a list called 'data'. Finally, we create a box plot of the data using the 'boxplot()' function, set the title and labels for the axes using the 'title()', 'xticks()', and 'ylabel()' functions, and display the plot using the 'show()' function.

### Running this code will produce a box plot of the two datasets side by side, with the boxes representing the IQR and the whiskers representing the rest of the distribution. The plot can be used to compare the distributions of the two datasets and identify any outliers.
