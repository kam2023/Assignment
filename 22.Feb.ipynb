```python
# URL: https://www.youtube.com/@PW-Foundation/videos
# Q1: Write a python program to extract the video URL of the first five videos.

## To extract the video URLs of the first five videos from the provided YouTube channel URL, you can use
## the 'requests' and 'beautifulsoup4' libraries in Python.

import requests
from bs4 import BeautifulSoup

# URL of the YouTube channel
url = 'https://www.youtube.com/@PW-Foundation/videos'

# Send a GET request to the URL and get the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the first five video links
video_links = soup.find_all('a', {'class': 'yt-uix-tile-link'})[:5]

# Extract the video URLs from the links
video_urls = ['https://www.youtube.com' + link['href'] for link in video_links]

# Print the video URLs
for url in video_urls:
    print(url)
    
##  This code extracts the video links from the HTML content using the 'find_all' method of BeautifulSoup, 
## which finds all HTML elements with the specified tag and attributes. In this case, we're looking for all links with the class
## 'yt-uix-tile-link', which are the video links on the channel page.
## We then extract the URLs from the links using a list comprehension and print them using a for loop. The '[:5]' slice 
## notation is used to limit the number of links to the first five.
```
